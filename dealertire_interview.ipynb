Something I have mixed feelings with. I’ll admit I completed a lot of this analysis and data exploration without dropping duplicate recall campaigns. It appears that each campaign and all subsequent information for the recall was duplicated, most importantly the number of potential vehicles affected. However, the year of the vehicle was the only aspect of each duplicate campaign that changed. As a manufacturer, it would make sense to count these as separate vehicles, but most would assume individual parts and components don’t change too much year over year. Also, the number of affect units being affected being exactly duplicated for each model seems pretty inaccurate. If the number of units affected differed between each model based on its make year, then it would make sense to add them up for a total for each campaign.  

For the sake of this analysis we’ll only look at each unique recall campaign. 

```python
recall_rawdata = recall_rawdata.drop_duplicates(subset='CAMPNO')
```

```python
recall_rawdata.shape
```

```python
recall_rawdata.corr()
```

```python
# seperate relevent vehicle/part manufactures from the rest of the data. 

bmw = recall_rawdata.query('MAKETXT == ["BMW"]')
mercedes = recall_rawdata.query('MAKETXT ==["MERCEDES BENZ"]')
mopar = recall_rawdata.query('MAKETXT ==["MOPAR"]')
lexus = recall_rawdata.query('MAKETXT ==["LEXUS"]')
nissan = recall_rawdata.query('MAKETXT ==["NISSAN"]')
audi = recall_rawdata.query('MAKETXT ==["AUDI"]')
hyundai = recall_rawdata.query('MAKETXT ==["HYUNDAI"]')
honda = recall_rawdata.query('MAKETXT == ["HONDA"]')
acura = recall_rawdata.query('MAKETXT == ["ACURA"]')
saab = recall_rawdata.query('MAKETXT == ["SAAB"]')
toyota = recall_rawdata.query('MAKETXT == ["TOYOTA"]')

# Combine into data set we will be working with.
recall_dt = pd.concat([bmw, mercedes, mopar, lexus, nissan, audi, hyundai, honda, acura, saab, toyota])

```

```python
recall_dt.head(3)
```

```python
recall_dt.describe()
```

```python
# significally cuts down on the number of recalls (probably a good thing!)

recall_dt.shape
```

```python
# Converting our date columns to pandas DateTime format for better analysis.

recall_dt['BGMAN'] = pd.to_datetime(recall_dt['BGMAN'].astype(str), format='%Y%m%d')
recall_dt['ENDMAN'] = pd.to_datetime(recall_dt['ENDMAN'].astype(str), format='%Y%m%d')
recall_dt['ODATE'] = pd.to_datetime(recall_dt['ODATE'].astype(str), format='%Y%m%d')
recall_dt['RCDATE'] = pd.to_datetime(recall_dt['RCDATE'].astype(str), format='%Y%m%d')
recall_dt['DATEA'] = pd.to_datetime(recall_dt['DATEA'].astype(str), format='%Y%m%d')
```

```python
# filter out some columns that we won't need.

recall_dt.drop(['RECORD_ID', 'MFGNAME','RPNO','FMVSS','DESC_DEFECT', 'CONEQUENCE_DEFECT','CORRECTIVE_ACTION','NOTES','RCL_CMPT_ID'], inplace=True, axis=1)
```

```python
# re-set index for the dataframe
# Setting the recall campaign begining date as our index for now. 
recall_dt = recall_dt.set_index(recall_dt.BGMAN)

recall_dt.head(3)
```

```python
parts_per_make = recall_dt.POTAFF.groupby(recall_dt.MAKETXT)
parts_affected = parts_per_make.sum().sort_values(ascending=False)

ax = parts_affected.plot(kind='bar', color='red')
ax.set_title("Number of vehicles affected per car make")
ax.set_ylabel("Total units affected")
ax.set_xlabel("Car make")
vals = ax.get_yticks()
ax.ticklabel_format(style='sci', axis='y')
```

```python
# Function to plot amount of vehicles affected for each recall per car make by model.

def car_model_plot(data):
    plt.figure(figsize=(10,7))

    car_model = data.POTAFF.groupby(data.MODELTXT)
    parts = car_model.sum().sort_values(ascending=False)[0:15]
    
    pl = parts.plot(kind='bar', color='purple')
    pl.set_title("Number of vehicles affected per model")
    pl.set_ylabel("Total units affected")
    pl.set_xlabel("Car Model")
    vals = pl.get_yticks()
    pl.ticklabel_format(style='sci', axis='y')

recall_dt.MAKETXT.unique() # use the individual dataframes we seperated from the raw data.
```

Add in each of the data frames created above (lowercase) and the funciton will plot the amount
of recalled parts for each car model.

**The number of recall campaigns over the years for each car make.**

Using the pandas pivot tabel can create a new data frame so if we wanted to view these metrics more indepth per make we could with this object. 


```python
recall_count_camp = recall_modern.pivot_table('CAMPNO', aggfunc=np.size, columns='MAKETXT',
index=recall_modern.index, fill_value=0)   

recall_count_camp.index = pd.DatetimeIndex(recall_count_camp.index)

plo = recall_count_camp.rolling(365).sum().plot(figsize=(20,30), subplots=True, layout=(-1,3), sharex=False, sharey=False)
```

**From this it looks that Nissan and Mercedes are the problem manufacturs for recalled parts in terms of the number of campaigs, lets see if its the same for the total amount of vehicles affected.**

**The amount of vehicles affected for each car make (average of vehicles affected per year)**

```python
recall_sum_parts = recall_modern.pivot_table('POTAFF', aggfunc=[np.sum],columns='MAKETXT', # could update aggfunc to np.mean for a more general trend.
index=recall_modern.index, fill_value=0)   

recall_sum_parts.index = pd.DatetimeIndex(recall_count_camp.index)

plo = recall_sum_parts.rolling(365).sum().plot(figsize=(18,30), subplots=True, layout=(-1,3), sharex=False, sharey=False)
```
